```python
headers = {
    'User-Agent': random.choice(USER_AGENTS),
}
asession = AsyncHTMLSession()
```

```python
product_list = []

for x in range (1, 5): 
    url = "https://www.lazada.sg/shop-mobiles/?clickTrackInfo=f7fd6c1a-e30e-439a-9ecf-2bc335bb14f5__3__2374658766__20__0.1__333258__7253__c2i__0.0__38%2525%2Boff&from=hp_categories&item_id=2374658766&params=%7B%22catIdLv1%22%3A%222%22%2C%22pvid%22%3A%22f7fd6c1a-e30e-439a-9ecf-2bc335bb14f5%22%2C%22src%22%3A%22ald%22%2C%22categoryName%22%3A%22Mobiles%22%2C%22categoryId%22%3A%223%22%7D&q=Mobiles&spm=a2o42.homepage.categoriesPC.d_1_3&src=hp_categories&up_id=2374658766&version=v2&page="
    response = await asession.get(url+ str(x))
    await response.html.arender()
    soup = BeautifulSoup(response.html.html, 'html.parser') 

    products = soup.find_all('div', class_ = "Bm3ON")
    print (products)
    

    for product in products: 
        print (product)
        # print (product) Correct data returned
        product_name = product.find("div", {'class' : "RfADt"})
        if product_name is not None:
            product_name = product_name.get_text("\\n").strip()
        price = product.find("span", class_ = "ooOxS").text
            
        discount = product.find("div", {"class" : "WNoq3"})
        if discount is not None: 
            discount = discount.get_text("\\n").strip()
            
        num_sold = product.find("span", {'class' : "_1cEkb"})
        if num_sold is not None: 
                num_sold = num_sold.get_text("\\n").strip()
                
        num_reviews = product.find("span", {'class' : "qzqFw"})
        if num_reviews is not None: 
            num_reviews = num_reviews.get_text().strip()
            
        location = product.find("span", {'class' : 'oa6ri'})
        if location is not None:
            location = location.get_text("\\n").strip()
            
        
        product_info = {
            'name' : product_name,
            'price' : price,
            'discount' : discount,
            'num_sold' : num_sold,
            'num_reviews' : num_reviews,
            'location' : location,
        }
        
        product_list.append(product_info)
        time.sleep(1)
```
